```python
import pandas as pd

df = pd.read_csv("mental_health_and_burnout.zip")
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
df = df.fillna(method='ffill')  # fills missing values from above
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


print(df.columns)
```

```python
sns.scatterplot(x='SleepHours', y='StressLevel', data=df)
plt.title("Stress vs Sleep Hours")
plt.xlabel("Sleep Hours")
plt.ylabel("Stress Level")
plt.show()
```

```python
sns.boxplot(x='HasMentalHealthSupport', y='BurnoutLevel', data = df)
plt.title("Burnout vs Mental Health Support")
plt.xlabel("Has Mental Health Support")
plt.ylabel("Burnout Level")
plt.show()
```
